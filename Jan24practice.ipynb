```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df1 = pd.read_csv("abalone.data")
df1.columns = ["Gender","Length","Diameter","Height","Whole-weight","Shucked-weight","Viscera-weight","Shell-weight","Rings"]
df1.head()
```

```python
genderList = df1["Gender"].tolist()
ls2 = []
for i in genderList:
    if(i=='F'):
        ls2.append(1)
    elif(i=='M'):
        ls2.append(2)
    else:
        ls2.append(3)
df1.drop(["Gender"],axis=1,inplace=True)
df1["Gender"] = ls2
df1.head()
```

# Relation between Gender of Abalones with the number of rings


```python
df_gender = df1.groupby("Gender")
df_gender.first()
```

# Relation between gender of abalones with their weights

```python
df_multipleWeightMax = df1.groupby("Gender")["Whole-weight","Shucked-weight","Viscera-weight","Shell-weight"].max()
df_multipleWeightMax
```

```python
df_multipleWeightMin = df1.groupby("Gender")["Whole-weight","Shucked-weight","Viscera-weight","Shell-weight"].min()
df_multipleWeightMin
```

```python
df_multipleWeightMean = df1.groupby("Gender")["Whole-weight","Shucked-weight","Viscera-weight","Shell-weight"].mean()
df_multipleWeightMean
```
