```python
#this function checks if the input password is valid or not
def password_checker(pswd):
    flag1 = 0
    flag2 = 0
    flag3 = 0
    flag4 = 0

    for char in pswd:
        if char.isupper():
            flag1 = 1
        elif char.islower():
            flag2 = 1  
        elif(char in '!@$#%&*^'):
            flag3 = 1 
        elif char.isdigit():
            flag4 = 1 

    if(flag1==1 and flag2==1 and flag3==1 and flag4==1 and len(pswd) >= 8):
        print('Valid Password')
    else:
        print('Invalid Password')
                
_pswd = input()
password_checker(_pswd)
```

```python
#THIS FUNCTION CALCULATES
def calculator(num1,num2,opr):
    if(opr == '+'):
        return num1+num2
    elif(opr == '-'):
        return num1-num2
    elif(opr == '*'):
        return num1*num2
    else:
        return num1/num2
    
num1 = int(input())
num2 = int(input())
opr = input()
calculator(num1,num2,opr)
```

```python
def fact(n):
    if(n==0):
        return 1
    elif(n==1):
        return 1
    else:
        return n*fact(n-1)
print(fact(6))
```
